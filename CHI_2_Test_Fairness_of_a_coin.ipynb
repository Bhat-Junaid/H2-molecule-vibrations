```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

N = 20
p = 0.5
x = np.arange(0, N+1)
def bi_dist(N, p):
    probabilities = [stats.binom.pmf(k, N, p) for k in x]
    return probabilities
prob = bi_dist(N, p)

print(prob)

plt.plot(x, prob)
plt.title(f"Binomial Distribution (N={N}, p={p})")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xticks(x)
plt.show()
```

```python
sum_prob = []
for i in range (len(prob)):
    if (i <= 8) or (i >= 12):
        sum_prob.append(prob[i])
print(sum(sum_prob))
```

#### Calculation of  $\chi^2 $ value and p-value

```python
f_obs = np.array([12, 8])
f_exp = np.array([10, 10])

chi_sq_value = stats.chisquare(f_obs, f_exp, ddof = 0)
print(chi_sq_value)
```

As seen from the above calculations the p-value > 0.05, therefore the hypothesis that the coin is fair can't be rejected.
